```python
driver = webdriver.Chrome(executable_path='../../chromedriver') # Could be any other browser you have the drivers for
driver.get(url)
html = driver.page_source
code = BeautifulSoup(html,'html5lib')
print code
```

```python
code.prettify()
```

```python
# function that takes a url and outputs the html on the url page

def extract_text(url):
    driver = webdriver.Chrome(executable_path='../../chromedriver') # Could be any other browser you have the drivers for
    driver.get(url)
    html = driver.page_source
    code = BeautifulSoup(html,'html5lib')
    print code.prettify()
```
